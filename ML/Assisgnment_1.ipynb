### Podpunkt 4

```python
Y3['class'].unique()
```

```python
best_cars = X3[np.logical_or(Y3['class'] == 'vgood', Y3['class'] == 'good')]
```

```python
for col in X3.columns:
    print(best_cars[col].value_counts())
    print('\n\n')
```

Moja zasada:
* Koszt auta: low, med
* Utrzymanie: czym nizsze tym lepsze
* Drzwi: nie maja znaczenia
* Ludzie: 4, more
* Bagaznik: czym wiekszy tym lepszy
* Bezpieczenstwo: high, med

```python
def classify(row):
    score = 0

    if row['buying'] in ['low', 'med']:
        score += 1
    if row['maint'] in ['low', 'med']:
        score += 1
    if row['persons'] in ['4', 'more']:
        score += 1
    if row['lug_boot'] in ['med', 'big']:
        score += 1
    if row['safety'] in ['high', 'med']:
        score += 1

    if score >= 3:
        return 'YES'
    else:
        return 'NO'

predictions = X3.apply(classify, axis=1)
```

```python
Y3['class'].unique()
```

```python
conv_dict = {
    'unacc': 'NO',
    'acc': 'NO',
    'good': 'YES',
    'vgood': 'YES'
}
```

```python
real = Y3['class'].map(conv_dict)
```

```python
TP = np.logical_and(predictions == "YES", real == 'YES')
TN = np.logical_and(predictions == "YES", real == 'NO')
FP = np.logical_and(predictions == "NO", real == 'YES')
FN = np.logical_and(predictions == "NO", real == 'NO')
```

```python
confusion_matrix = np.array([[TP.sum(), FN.sum()],
                              [FP.sum(), TN.sum()]])

plt.figure(figsize=(8, 6))
plt.imshow(confusion_matrix, cmap='Blues')
plt.colorbar()

for i in range(2):
    for j in range(2):
        plt.text(j, i, confusion_matrix[i, j], ha='center', va='center')

plt.xticks([0, 1], ['Predicted YES', 'Predicted NO'])
plt.yticks([0, 1], ['Actual YES', 'Actual NO'])
plt.title('Confusion Matrix')
plt.show()
```
